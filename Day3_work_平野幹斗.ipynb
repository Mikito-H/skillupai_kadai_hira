```python
# 係数を求める
y = data["distance"].values
X = data[["speed","consume"]].values
regr = LinearRegression(fit_intercept=True)
regr.fit(X, y)

# 重みを取り出す
w0 = regr.intercept_
w1 = regr.coef_[0]
w2 = regr.coef_[1]
```

```python
# 変数の設定
x1 = data["speed"]
x2 = data["consume"]
y = data["distance"]

# 重みと二乗誤差の確認
y_est = w0 + w1 * x1 + w2 * x2
squared_error = 0.5 * np.sum((y - y_est) ** 2)
print('w0 = {:.3f}, w1 = {:.3f}, w2 = {:.3f}, 二乗誤差 = {:.3f}'.format(w0, w1, w2, squared_error))

# 作図の設定
fig = plt.figure()
ax = Axes3D(fig)

# 軸の設定
ax.set_xlabel("speed")
ax.set_ylabel("consume")
ax.set_zlabel("distance")

# プロット
ax.scatter3D(x1, x2 ,y)

# 直線のプロット
x1 = np.arange(min(x1), max(x1), (max(x1) - min(x1)) / 100)
x2 = np.arange(min(x2), max(x2), (max(x2) - min(x2)) / 100)
x1, x2 = np.meshgrid(x1, x2)
y_est = w0 + w1 * x1 + w2 * x2
ax.plot_wireframe(x1, x2, y_est, color="orange")
plt.show()
```
